**5. Perform meaningful visualizations using the dataset.**

*a)Histogram*


```python
plt.hist(data['Glucose'],bins=30)
plt.title('Histogram of Glucose')
plt.xlabel('Glucose Value')
plt.ylabel('Frequency')
plt.show()
```

*b)Boxplot*

```python
# sns.boxplot(x=data['BMI'])
plt.boxplot(data['BMI'])
plt.title('Box Plot of BMI')
plt.show()
```

*c)Scatterplot*

```python
# plt.plot(data['Glucose'], data['BloodPressure'])
sns.scatterplot(data=data, x='Glucose', y='BloodPressure',alpha=.75)
plt.title('Scatter Plot of Glucose vs BloodPressure')
plt.show()
```

*d)Heatmap*

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
